# 'SUMAS DE RIEMANN'

Es un tipo de aproximación del valor de una integral mediante una suma finita, La suma se calcula dividiendo la región en formas (rectángulos, trapezoides, cuadrados, triángulo, parábolas o cúbicas) que juntas forman una región que es similar a la región que se está midiendo, luego calculando el área para cada una de estas formas y, finalmente, agregando todas estas pequeñas áreas juntas.

##'Sumas de Riemann por la derecha

* Una suma de Riemann derecha usa rectángulos cuyos vértices superiores derechos están sobre la curva. En la gráfica de la función
la región bajo la curva está dividida en 4 rectángulos con el mismo ancho, tocando la curva en las esquina superiores izquierdas

Si yi = xi, entonces denominamos S como la suma de Riemann por la derecha.

utilizando  la siguiente formula: $$ S = \sum_{i=1}^{n}(f(x)(X{i} - X{i} - 1)$$

# **Paso 1: Determinar la partición**

```

def F_particionder(a, b, n):

   delta = (b - a) / n
   particionder = [ ]

   for i in range(1, n+1, 1):           # Para la suma de Riemann derecha es range(1, n+1, 1)
     x = a + i * (delta)              # Para las sumas de Riemann intermedias x = (a + delta / 2) + i * delta.
     particionder.append(x)
   return particionder

F_particionder(0, 2, 3) 
```

# **Paso 2: Evaluar los elementos de la partición en la función (Hallar las alturas)**

```
def F_evalpart(particionder):

  # Creamos una lista vacía
  evalpart = []

  # Creamos un for para que evalue cada uno de los elementos de la partionder en la función
  for i in range(0,len(particionder),1):
    x = particionder[i]
    y = (x ** 3)
    evalpart.append(y)
    #print(y)
  return evalpart 

#Llamamos la función F_evalpart.
F_evalpart(F_particionder(0, 2, 3)) 
```

# **Paso 3: Multiplicar delta por las alturas - hallar las áreas de los rectangulos**

```
def F_areasrect(evalpart, a, b, n):
  
  # Creamos una lista vacía para que acumule las áreas de los rectangulos
  areasrect = []

  # Generar delta nuevamente
  delta = (b - a) /n 

  # Creamos un for para que vaya multiplicando las alturas por delta para generar las áreas de los rectangulos
  for i in evalpart:
    x = i * delta 
    areasrect.append(x)
  return areasrect

#Llamamos la funcion F_areasrect para confirmar que todo vaya bien.
F_areasrect(F_evalpart(F_particionder(0, 2, 3)), 0, 2, 3)
```

# **Paso 4: Sumar las áreas de los rectángulos**

```
def F_sumaRiemann(areasrect):
  # Creamos una variable igual a cero para que almacene la suma
  sumaRiemman = 0

  for i in areasrect:
    sumaRiemman = sumaRiemman + i
    #           =             +
  return sumaRiemman

# Llamamos la función F_sumaRiemann para confirmar que todo vaya bien.
F_sumaRiemann(F_areasrect(F_evalpart(F_particionder(0, 2, 3)), 0, 2, 3))  
```

La siguiente solución se plantea a partir de la definición de las sumas de Riemann usando la sumatoria, donde A es el área bajo la curva de la función en el intervalo dado  $$ Aca - la - sumatoria $$

```
def SumaRapidaRiemann(a, b, n):

  summR = 0
  for i in range(0, n, 1):
    delta = (b - a) / n               #delta
    x = a + (i * delta)               #Paso1
    y = (x ** 3)                      #Paso2
    suma = y * delta                  #Paso3
    summR = summR + suma              #Paso4

  return summR

print("La suma de riemann es: ", SumaRapidaRiemann(0, 2, 3))
```
