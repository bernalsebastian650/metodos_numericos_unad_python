# **Paso 1: Determinar la partici칩n**

```

def F_particionder(a, b, n):

   delta = (b - a) / n
   particionder = [ ]

   for i in range(1, n+1, 1):           # Para la suma de Riemann derecha es range(1, n+1, 1)
     x = a + i * (delta)              # Para las sumas de Riemann intermedias x = (a + delta / 2) + i * delta.
     particionder.append(x)
   return particionder

F_particionder(0, 2, 3) 
```

# **Paso 2: Evaluar los elementos de la partici칩n en la funci칩n (Hallar las alturas)**

```
def F_evalpart(particionder):

  # Creamos una lista vac칤a
  evalpart = []

  # Creamos un for para que evalue cada uno de los elementos de la partionder en la funci칩n
  for i in range(0,len(particionder),1):
    x = particionder[i]
    y = (x ** 3)
    evalpart.append(y)
    #print(y)
  return evalpart 

#Llamamos la funci칩n F_evalpart.
F_evalpart(F_particionder(0, 2, 3)) 
```

# **Paso 3: Multiplicar delta por las alturas - hallar las 치reas de los rectangulos**

```
def F_areasrect(evalpart, a, b, n):
  
  # Creamos una lista vac칤a para que acumule las 치reas de los rectangulos
  areasrect = []

  # Generar delta nuevamente
  delta = (b - a) /n 

  # Creamos un for para que vaya multiplicando las alturas por delta para generar las 치reas de los rectangulos
  for i in evalpart:
    x = i * delta 
    areasrect.append(x)
  return areasrect

#Llamamos la funcion F_areasrect para confirmar que todo vaya bien.
F_areasrect(F_evalpart(F_particionder(0, 2, 3)), 0, 2, 3)
```

# **Paso 4: Sumar las 치reas de los rect치ngulos**

```
def F_sumaRiemann(areasrect):
  # Creamos una variable igual a cero para que almacene la suma
  sumaRiemman = 0

  for i in areasrect:
    sumaRiemman = sumaRiemman + i
    #           =             +
  return sumaRiemman

# Llamamos la funci칩n F_sumaRiemann para confirmar que todo vaya bien.
F_sumaRiemann(F_areasrect(F_evalpart(F_particionder(0, 2, 3)), 0, 2, 3))  
```

La siguiente soluci칩n se plantea a partir de la definici칩n de las sumas de Riemann usando la sumatoria, donde A es el 치rea bajo la curva de la funci칩n en el intervalo dado  $$ Aca - la - sumatoria $$

```
def SumaRapidaRiemann(a, b, n):

  summR = 0
  for i in range(0, n, 1):
    delta = (b - a) / n               #delta
    x = a + (i * delta)               #Paso1
    y = (x ** 3)                      #Paso2
    suma = y * delta                  #Paso3
    summR = summR + suma              #Paso4

  return summR

print("La suma de riemann es: ", SumaRapidaRiemann(0, 2, 3))
```
