# Paso 1: Definimos la función 



```python
import numpy as np

def f(x):
   return 1 / np.cos(x)/(1/np.tan(x)) 
```

# Definimos derivadas

```python
def fpf(x0, h): # esquema primera derivada hacia delante
    return (f(x0 + h) -f(x0)) / h
```

```python
def fpc(x0, h):# esquema primera derivada hacia atras
    return (f(x0 + h) - f(x0 - h)) / (2 * h)
```

```python
def fpb(x0, h): # esquema primera derivada central
    return (f(x0) - f(x0 - h)) / h
```

# Paso 3: Definimos Datos, ejecución derivadas

```python
x = (27.5 * np.pi) / 180
h = 0.5
print("El valor de x en radianes es:", x)
```

```python
fpb (x,h) # Esquema primera derivada hacia central
```

# Tabla derivada Adelante

```python
DR = (np.cos(x) )  # derivada de sen(), primer elemento
```

```python
def ErrorRP(DR, x, h):     # Definimos el error relativo, segundo elemento
  return(abs(DR - fpf(x, h)) / DR) * 100 
```

```python
from tabulate import tabulate

DFinitas = [["0.5", fpf(x, 0.5), ErrorRP(DR, x, 0.5)], ["0.05", fpf(x, 0.05), ErrorRP(DR, x, 0.05)],["0.01", fpf(x, 0.01), ErrorRP(DR, x, 0.01)]]
print(tabulate(DFinitas, headers = ["tamaño h", "Derivada aprox", "Error(%)"], tablefmt = "francy_grid"))
```

# Segunda Derivada

# Definimos esquema segunda derivada 

```python
def fsf(x0, h): # esquema segunda derivada hacia delante
    return (f(x0) - 2 * f(x0 + h) + f(x0 + 2 * h)) / h ** 2
```

```python
def fsc(x0, h): # esquema segunda derivada hacia atras
    return (f(x0 - h) - 2 *(f(x0) + f(x0) + h)) / h ** 2
```

```python
def fsb(x0, h): # esquema segunda derivada central
    return (f(x0 - 2 * h) - 2 * f(x0 - h) + f(x0)) / h ** 2
```

# Tabla para la segunda derivada hacia adelante

```python
SDR =-(np.sin(x)) # definimos la segunda derivada
```

```python
def ErrorRP(SDR, x, h):  # definimos el error relativo porcentual
  return(abs(SDR - fsf(x, h)) / SDR) * 100   
```

```python
DFinitas = [["0.5", fsf(x, 0.5), ErrorRP(DR, x, 0.5)], ["0.05", fsf(x, 0.05), ErrorRP(DR, x, 0.05)],["0.01", fsf(x, 0.01), ErrorRP(DR, x, 0.01)]]
print(tabulate(DFinitas, headers = ["tamaño h", "Derivada aprox", "Error(%)"], tablefmt = "francy_grid")) # se pinta la tabla para segunda derivada hacia delante
```
