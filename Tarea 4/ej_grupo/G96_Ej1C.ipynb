$$f'(x_{0}) = \frac{f(x_{0}+h)-f(x_{0})}{h}$$

# Diferencias Finitas - Primera Derivada

**Paso 1: Importamos Numpy y definimos la función**


```python
import numpy as np

def f(x):
   return np.sin(x)/np.cos(x)                                 # Se agrega la función que se desea trabajar
   
```

**Paso 2: Definimos nuestros esquemas de derivadas**


```python
# Esquema df primera derivada Foward
# f prima Foward

def fpf(x0, h):
    return (f(x0 + h) -f(x0)) / h
```

```python
# Esquema df primera derivada central

def fpc(x0, h):
    return (f(x0 + h) - f(x0 - h)) / (2 * h)
```

```python
# Esquema df primera derivada Backward

def fpb(x0, h):
    return (f(x0) - f(x0 - h)) / h
    
```

image.png

**Paso 3: Ingreso de datos y ejecución de las derivadas**

```python
x = (27.5 * np.pi) / 180
h = 0.5
print("El valor de x en radianes es:", x)
```

```python
fpf(x, h)                                # Esta es a primera derivada de la funcion tan(x) hacia adelante
```

```python
fpc(x, h)                                # Esta es la primera derivada de la función tan(x) central
```

```python
fpb(x, h)                                # Esta es la primera derivada de la función tan(x) hacia atras
```

**Paso 4.1: Tabla de la primera derivada hacía adelante**

**Elemento 1: Definimos la derivada real**

```python
# Definimos la derivada y guardamos el valor en  la variable DR

DR = (np.sin(x)) / (np.cos(x))                                    # Agregamos la derivada analítica
```

**Elemento 2: El error relativo porcentual**

```python
def ErrorRP(DR, x, h):                                            # Definimos el error relativo porcentual
  return(abs(DR - fpf(x, h)) / DR) * 100                          # Utilizamos el esquema de la primera derivada hacia adelante
```

```python
from tabulate import tabulate

DFinitas = [["0.5", fpf(x, 0.5), ErrorRP(DR, x, 0.5)], ["0.05", fpf(x, 0.05), ErrorRP(DR, x, 0.05)],["0.01", fpf(x, 0.01), ErrorRP(DR, x, 0.01)]]
print(tabulate(DFinitas, headers = ["tamaño h", "Derivada aprox", "Error(%)"], tablefmt = "francy_grid"))
```

ANALISÍS: Después de revisar los resultados arrojados para la tabla de la primera derivada hacía adelante podemos afirmar que entre mayor sea el tamaño de h el error relativo porcentual es mayor para el caso de 0.5 arroja un error de 272.778 y el tamaño 0.01 arroja 145.441

**Paso 4.2 - Tabla para primera derivada hacia atras**

```python
def ErrorRP(DR, x, h):                                                      # Definimos el error relativo porcentual
  return(abs(DR - fpb(x, h)) / DR) * 100                                    # Utilizamos el esquema de la primera derivada hacia atras
```

```python
from tabulate import tabulate

DFinitas = [["0.5", fpb(x, 0.5), ErrorRP(DR, x, 0.5)], ["0.05", fpb(x, 0.05), ErrorRP(DR, x, 0.05)],["0.01", fpb(x, 0.01), ErrorRP(DR, x, 0.01)]]
print(tabulate(DFinitas, headers = ["tamaño h", "Derivada aprox", "Error(%)"], tablefmt = "francy_grid"))
```

ANALISÍS: Para la tabla de la primera derivada hacía atras decimos que entre mayor sea el tamaño de h el error relativo porcentual es menor, para el caso de 0.5 arroja un error de 107.698 y el tamaño 0.01 arroja 142.899

**Paso 4.2 - Tabla para primera derivada centrada**

```python
def ErrorRP(DR, x, h):                                                          # Definimos el error relativo porcentual
  return(abs(DR - fpc(x, h)) / DR) * 100                                        # Utilizamos el esquema de la primera derivada centrada
```

```python
from tabulate import tabulate

DFinitas = [["0.5", fpc(x, 0.5), ErrorRP(DR, x, 0.5)], ["0.05", fpc(x, 0.05), ErrorRP(DR, x, 0.05)],["0.01", fpc(x, 0.01), ErrorRP(DR, x, 0.01)]]
print(tabulate(DFinitas, headers = ["tamaño h", "Derivada aprox", "Error(%)"], tablefmt = "francy_grid"))
```

# $ **Diferencias Finitas - Para la Segunda Derivada**$

**Paso 1: Definimos Nuestros Esquemas para la Segunda Derivada**

```python
# Esquema df seguda derivada de Foward

def fsf(x0, h):
    return (f(x0) - 2 * f(x0 + h) + f(x0 + 2 * h)) / h ** 2
```

```python
# Esquema df segunda derivada central

def fsc(x0, h):
    return (f(x0 - h) - 2 *(f(x0) + f(x0) + h)) / h ** 2
```

```python
# Esquema df segunda derivada Backward

def fsb(x0, h):
    return (f(x0 - 2 * h) - 2 * f(x0 - h) + f(x0)) / h ** 2
```

image.png

**Paso 2.1: Tabla de la segunda derivada hacia adelante**

```python
# Definimos la segunda derivada real

SDR = 2 * (np.sin(x)) / (np.cos(x)) ** 3                # Definimos la segunda derivada
```

```python
def ErrorRP(SDR, x, h):                                                          # Definimos el error relativo porcentual
  return(abs(SDR - fsf(x, h)) / SDR) * 100                                       # Utilizamos el esquema de la segunda derivada hacia adelante
```

```python
DFinitas = [["0.5", fsf(x, 0.5), ErrorRP(DR, x, 0.5)], ["0.05", fsf(x, 0.05), ErrorRP(DR, x, 0.05)],["0.01", fsf(x, 0.01), ErrorRP(DR, x, 0.01)]]
print(tabulate(DFinitas, headers = ["tamaño h", "Derivada aprox", "Error(%)"], tablefmt = "francy_grid"))
```

ANALISÍS: Después de revisar los resultados arrojados para la tabla de la segunda derivada hacía adelante podemos afirmar que entre mayor sea el tamaño de h el error relativo porcentual es mayor para el caso de 0.5 arroja un error de 6445.22 y el tamaño 0.01 arroja 163.221

**Paso 2.2: Tabla para segunda derivada hacia atras**

```python
def ErrorRP(SDR, x, h):                                                           # Definimos el error relativo porcentual
  return(abs(SDR - fsb(x, h)) / SDR) * 100                                        # Utilizamos el esquema de la segunda derivada hacia atras
```

```python
DFinitas = [["0.5", fsb(x, 0.5), ErrorRP(DR, x, 0.5)], ["0.05", fsb(x, 0.05), ErrorRP(DR, x, 0.05)],["0.01", fsb(x, 0.01), ErrorRP(DR, x, 0.01)]]
print(tabulate(DFinitas, headers = ["tamaño h", "Derivada aprox", "Error(%)"], tablefmt = "francy_grid"))
```

ANALISÍS: Para la tabla de la primera derivada hacía atras decimos que entre mayor sea el tamaño de h el error relativo porcentual es menor, para el caso de 0.5 arroja un error de 109.195 y el tamaño 0.01 arroja 145.509

**Paso 2.3: Tabla par la segunda derivada central**
