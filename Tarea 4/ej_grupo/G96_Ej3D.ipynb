#**Paso 1: Definimos la función**


```python
import matplotlib.pyplot as plt                  # Importamos librerias para gráficar 
import numpy as np                               # Importamos librerias numpy
import scipy.special as sp

def fun(x, y):                                   # Definimos la función fun
  return y**2 / (x + y**2)
```

#**Paso 2 : Método Runge – Kutta 1er orden – Euler**

Los métodos de Runge-Kutta (RK) logran una exactitud del procedimiento de una
serie de Taylor, sin requerir el cálculo de derivadas superiores.
Probablemente uno de los procedimientos más difundidos, y a la vez más exactos,
para obtener la solución numérica del problema de valor inicial:
$y´ = f(t,y)$, con   $y(to) = yo$, sea el método de Runge- Kutta de cuarto orden.
Los métodos de Runge Kutta de cualquier orden se deducen mediante el desarrollo
de la serie de Taylor de la función $f(t,y).$

El procedimiento de Runge-Kutta se puede programar fácilmente en los ordenadores, y además se emplea mucho en la práctica, debido a la su exactitud relativamente elevada de la solución aproximada de la ecuación diferencial

```python
# dy/dx = f(x,y) en el intervalo de [0, 5] con y(0) = 14
# h=0.5 es el tamaño de paso 

def runge_kutta_1(f, a, b, h, y0):
  x = [a]
  y = [y0]
  while x [-1] < b:
    y.append(y[-1] + f(x[-1], y[-1]) * h)
    x.append(x[-1] + h)
  return x, y
```

#**Paso 3 : Método Runge – Kutta 2do orden** 
Una ecuación diferencial de segundo orden es equivalente a un sistema de dos ecuaciones diferenciales de primer orden

```python
# Método Runge – Kutta 2do orden para h= 0.5

def runge_kutta_2(f, a, b, h, y0):
  x = [a]
  y = [y0]
  while x [-1] < b:
    k1 = f(x[-1], y[-1])
    k2 = f(x[-1] + 3 / 4 * h, y[-1] + 3 / 4 * k1 * h)
    y.append(y[-1] + (1 / 3 * k1 + 2 / 3 * k2) * h)
    x.append(x[-1] + h)
  return x, y
```

#**Paso 4 : Método Runge – Kutta 4to orden** 
El método Runge-Kutta de orden 4 es la forma de los métodos de Runge-Kutta de uso mas común y así mismo mas exactos para obtener soluciones aproximadas de ecuaciones diferenciales. La solución que ofrece este método, es una tabla de la función solución, con valores de “y” correspondientes a valores específicos de “x”.


```python
# Método Runge – Kutta 4to orden para h=0.5

def runge_kutta_4(f, a, b, h, y0):
  x = [a]
  y = [y0]
  while x [-1] < b:
    k1 = f(x[-1], y[-1])
    k2 = f(x[-1] + 0.5 * h, y[-1] + 0.5 * k1 * h)
    k3 = f(x[-1] + 0.5 * h, y[-1] + 0.5 * k2 * h)
    k4 = f(x[-1] + h, y[-1] + k3 * h)
    y.append(y[-1] + (h / 6) * (k1 + 2 * k2 + 2 * k3 + 4))
    x.append(x[-1] + h)
  return x, y
```

```python
h = 0.5
print("Tamaño de paso(h):" , h)
x1, y1 = runge_kutta_1(fun, 0, 4, h, 14)
x2, y2 = runge_kutta_2(fun, 0, 4, h, 14)
x4, y4 = runge_kutta_4(fun, 0, 4, h, 14)
#x5 = np.arange(0, 4, 0.01)
#y5 = y'=y^2/(x+y^2),y(0)=1  # Solución analítica de la ED

```
