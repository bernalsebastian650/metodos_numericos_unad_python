#**Ejercicio 3: Ecuaciones Diferenciales Ordinarias**




#**Paso 1: Definimos la función**


```python
import matplotlib.pyplot as plt 
import numpy as np

def fun(x,y):
  return y*pow(x,2) + y
```

#**Paso 2 : Resuelvemos la EDO por el método Runge – Kutta 1er orden – Euler**

```python
def runge_kutta_1(f,a,b,h,y0):
  x=[a]
  y=[y0]
  while x[-1]<b:
    y.append(y[-1]+f(x[-1],y[-1])*h)
    x.append(x[-1]+h)
    return x,y
```

#**Paso 3 : Resuelvemos la EDO por el método Runge – Kutta 2do orden** 

```python
def runge_kutta_2(f,a,b,h,y0):
  x=[a]
  y=[y0]
  while x[-1]<b:
    k1=f(x[-1],y[-1])
    k2=f(x[-1]+3/4*h,y[-1]+3/4*k1*h)
    y.append(y[-1]+(1/3*k1+2/3*k2)*h)
    x.append(x[-1]+h)
  return x,y
```

#**Paso 4 : Resuelvemos la EDO por el método Runge – Kutta 4to orden** 

```python
def runge_kutta_4(f,a,b,h,y0):
  x=[a]
  y=[y0]
  while x[-1]<b:
    k1=f(x[-1],y[-1])
    k2=f(x[-1]+0.5+h,y[-1]+0.5*k1*h)
    k3=f(x[-1]+0.5*h,y[-1]+0.5*k2*h)
    k4=f(x[-1]+h,y[-1]+k3+h)
    y.append(y[-1]+(h/6)*(k1+2*k2+2*k3+k4))
    x.append(x[-1]+h)
  return x,y


```

#**Paso 5: Graficamos** 

```python
h=0.5
print("tamaño de paso(h):",h)
x1,y1=runge_kutta_1(fun,0,1,h,2)
x2,y2=runge_kutta_2(fun,0,1,h,2)
x4,y4=runge_kutta_4(fun,0,1,h,2)

plt.figure(figsize=(16, 10), dpi=80)
plt.plot(x1,y1,label="RK 1er orden")
plt.plot(x2,y2,label="RK 2do orden")
plt.plot(x4,y4,label="RK 4to orden")
plt.legend()
plt.show()
```
