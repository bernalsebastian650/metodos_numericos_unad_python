```python
n = len(b)               # n es igual a la longitud del vector b                                                    
count = 0                # contador igual a 0                                                   
for i in range(n):        # Recorremos la matriz de i hasta n                                                  
    Aii = A[i][i]          #  Inicializamos la posición Aii  y buscamos en la lista A[i][i]                                           
    suma = 0                 # inicicializamos la variable Suma en 0                                         
    for j in range(n):           # Recorremos la matriz j hasta n                                       
        if i != j:                       #Condición de que si i es distinto de j realizala suma                                 
           suma = suma + abs(A[i][i])     #   Sumamos los valores absolutos en la lista A[i][i]                             
    if Aii > suma:                           # condición de comparar que si Aii es mayor  la suma es Diagonal Dominante                         
       print("La fila", i, "es Diagonal Dominante")                               
       count += 1                              #  condición de comparar "entonces" Aii es menor a la suma No es Diagonal Dominante                            
    else:                                                                   
       print("La fila", i, "No es Diagonal Dominante")                                      
       break                                     # Cerrar el ciclo                            
if count == n:                                      #    Condición cuando el contador termine de leer la matriz imprime los siguientes mensajes                  
    print("La matriz es Dieganonal Dominante y el método iterativo convergira.")              
else:                                                                       
    print("La matriz no es Dieganonal Dominante y el método iterativo puede que no converja") 
```

# Metodo Jacobi

Es un método iterativo, usado para resolver sistemas de ecuaciones lineales del tipo. El algoritmo toma su nombre del matemático alemán Carl Gustav Jakob Jacobi. El método de Jacobi consiste en usar fórmulas como iteración de punto fijo.

```python
tol = 0.000001         # tolerancia del error                              
error = 100000         #  valor del error                                          
nitermax = 100         #  maxima tolerancia                          
n = len(A)               # n igual a la longitud de A                       
x = np.zeros(n)           # Vector inicial                           
xl = np.zeros(n)           # Vector auxiliar                        
niter = 0                   #Contador inical                       
Niter = []
Error = []


while niter < nitermax and error > tol: # mientras contador incial menor a maximo tolerancia y al error  mayor a tolerancia temine el ciclo
    for i in range(n):     #   Recorremos de i hata n                     
        d = b[i]            # d igual a la posición b[i]
        for j in range(n):   #    recorremos j hasta n                    
            if i != j:        # condición de que si i es distinto de j
              d = d - A[i][j] * x[j] # d resta a[i][j] del metodo jacobi
        x[i] = d / A[i][i]
    
    error = np.linalg.norm(x-xl)      # Calculo de la norma de calculo              
    xl = np.copy(x)                   # Actualiza el vector               
    print("Iteración:", niter)
    print(x)
    print("Error en la iteración", niter, "es:", error)

    niter += 1      # contador +1 hasta que llegue a nitermax
    Niter.append(niter)
    Error.append(error)
```

# Programación de Gauss Seidel

```python
def SOR(A,b, x, imax, ea, omega):             # Definimos la funcion de Gauss Seidel con los parametros
    n = len(x)                                
    for i in range(n):                        # Recorremos la función de i hasta n
        dummy = A[i][i]                        # dummy es igual a la matriz A[i][i]
        for j in range(n):                     # Recorremos la función de j hasta n
            A[i][j] = A[i][j] / dummy         # Asigignamos la matriz A[i][i] y la divide con dummy

        b[i] = b[i] / dummy                  # asignamos el vector b de cada valor y lo divide con dummy

    for i in range(n):                     # Recorremos el for para asignar a la variable sum el vector b y la matriz A[i][i]
        sum = b[i]
        for j in range(n):                    
            if i != j:
              sum -= A[i][j] * x[j]
        x[i] = sum                            

    iter = 1                                  

    while iter <= imax or ea > tol:           # Mientras la variable iteración sea mayor o igual al valor maximo o valor del error sea mayor a la tolerancia 
        centinela = 1                           
        for i in range(n):                    
            old = x[i]
            sum = b[i]
            for j in range(n):                
                if i!=j:
                    sum -= A[i][j] * x[j]     

                x[i] = omega * sum + (1- omega) * old  
                if centinela == 1 and x[i] != 0:
                   ea = abs((x[i] - old) / x[i]) * 100
                   if ea > tol:
                      centinela = 0

            print(iter)                     
            print(x)                        
            iter += 1                       
            if centinela == 1 or iter >= imax: 
                break
    return x  
```
