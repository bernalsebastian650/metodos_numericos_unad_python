# **METODO JACOBI**

Es un método iterativo, usado para resolver sistemas de ecuaciones lineales del tipo $$ {\displaystyle A\mathbf {x} =\mathbf {b}}. $$ El algoritmo toma su nombre del matemático alemán Carl Gustav Jakob Jacobi. El método de Jacobi consiste en usar fórmulas como iteración de punto fijo.

Su formula general es: $$ x_{i}^{(k)} = -{\frac{\ 1}{a_{ii}}= \{\sum_{j=i}^{n}a_{ij}x_{j}^{(k-1)}}\} + \frac{\ b_{i}}{a_{ii}} $$ , para cada i, 1<=i<=n

* Comprobación de diagonal dominante de la matriz A

En matemáticas, y concretamente en álgebra lineal, una matriz es de diagonal estrictamente dominante, cuando lo es por filas o por columnas.

* Lo es por filas cuando, para todas las filas, el valor absoluto del elemento de la diagonal de esa fila es estrictamente mayor que la norma del resto de elementos de esa fila.
* Lo es por columnas cuando, para todas las columnas, el valor absoluto del elemento de la diagonal de esa columna es estrictamente mayor que la norma del resto de elementos de esa columna.

```python
n = len(b)                                                                  # n es igual a la longitus del vector b
count = 0                                                                   # Contador que inicia en 0
for i in range(n):                                                          # para el rango de i hasta 10
    Aii = A[i][i]                                                           # busca la posicion en la matrz
    suma = 0                                                                # variable suma que arranca en cero
    for j in range(n):                                                      # para el rango de j hasta 10
        if i != j:                                                          # Si i es diferente de j 
           suma = suma + abs(A[i][i])                                       # realiza la suma absoluta de los componentes de la fila
    if Aii > suma:                                                          # si Aii es mayor que la varible suma
       print("La fila", i, "es DD")                                         # imprima la fila i es diagonal dominante
       count += 1                                                           # Almacena el resultado en un contador e inicia nuevamente el proceso
    else:                                                                   # Si no
       print("La fila", i, "No es DD")                                      # imprima la fila i no es diagonal dominante
       break                                                                # para salir de la condición
if count == n:                                                              # si el contador es igual a 10
    print("La matriz es DD y el método iterativo convergira.")              # imprima la matriz es diagonal dominante y el método iterativo convergira
else:                                                                       # si no
    print("La matriz no es DD y el método iterativo puede que no converja") # imprima la matriz no es diagonal dominante y el método iterativo puede que no converja.
  
```

* **Programación del método de jacobi**

El método de Jacobi permite hallar las aproximaciones a una solución de sistemas de ecuaciones lineales, utilizando los valores iniciales para la primera aproximación, luego los de la primera para la segunda y así sucesivamente; en este método el cálculo de cada variable es independiente por lo tanto ninguna variable depende de la otra.

El procedimiento a seguir para la aplicación del método es el siguiente:

* Se debe introducir unas aproximaciones iniciales, la matriz de coeficientes, el vector de términos independientes, una tolerancia y un número total de iteraciones. 
* Se toman las aproximaciones iniciales para hallar las nuevas aproximaciones, teniendo en cuenta el fundamento del método. 
* En cada paso, es posible calcular el error, que es este caso está definido en normas (las cuales son infinitas).
* Para la finalización del programas se tiene en cuenta, si el programa sobrepasa el numero de iteraciones, o si el error es menor del 

propuesto al principio; una vez ocurra alguna de estas dos situaciones, la ultima iteración tendrá las aproximaciones a la solución del sistema de ecuaciones estudiado.

```python
# Valores iniciales

tol = 0.000001                                       # Valor de la tolerancia de error
error = 100000                                       # Valor inicial de la norma                    De error = 3          error = 16
nitermax = 10                                       # Número máximo de iteraciones
n = len(A)                                           # Determinador del tamaño de A
x = np.zeros(n)                                      # Vector inicial de ceros
xl = np.zeros(n)                                     # Vector auxiliar de ceros
niter = 0                                            #  Contador inicial de iteración
Niter = []
Error = []


while niter < nitermax and error > tol:
    for i in range(n):                               # i es para recorrer las columnas
        d = b[i]
        for j in range(n):                           # j para recorrer las filas 
            if i != j:
              d = d - A[i][j] * x[j]
        x[i] = d / A[i][i]
    
    error = np.linalg.norm(x-xl)                     # Calculo de la norma vectorial
    xl = np.copy(x)                                  # Actualiza vector solución iteración anterior
    
    print("Iteración:", niter)
    print(x)
    print("Error en la iteración", niter, "es:", error)

    niter += 1
    Niter.append(niter)
    Error.append(error)
```

# $$ **PASO 2.1** $$

```python
def SOR(A,b, x, imax, ea, omega):             # Se define la función SOR con sus respectivos parámetros de entrada
    n = len(x)                                # La variable n tendrá el tamaño del vector de ceros X
    for i in range(n):                        # Se utiliza la primera función for para asignar cada valor de la matriz A a la variable dummy
        dummy = A[i][i]         
        for j in range(n):                    # El segundo for vuelve a recorrer cada valor de la matriz A y lo divide en sí mismo
            A[i][j] = A[i][j] / dummy

        b[i] = b[i] / dummy                   # Se recorre el vector B y cada valor de este es divido por el valor que tiene la variable dummy

    for i in range(n):                        # Se declara una función for para asignar los nuevos valores del vector B a la variable sum
        sum = b[i]
        for j in range(n):                    # Se declara un for para ir sustrayendo el valor actual de la matriz A en la posición i,j por el valor del vector X en la posición j cuando la coordenada de la matriz no corresponde a la diagonal principal y se le asigna a la variable sum
            if i != j:
              sum -= A[i][j] * x[j]
        x[i] = sum                            # Por cada recorrido del for, se actualiza el dato de la variable sum en la posición i del vector X

    iter = 1                                  # La variable de iteraciones se le asigna 1 

    while iter <= imax or ea > tol:           # Mientras la iteración sea menor o igual al numero deseado de iteraciones o mientras el error sea mayor que la tolerancia definida
        centinela = 1
        for i in range(n):                    # Se asignan los valores actuales del vector x en la variable old y del vector b en la variable sum
            old = x[i]
            sum = b[i]
            for j in range(n):                #Se declara una función for para asignar los nuevos valores del vector B a la variable sum
                if i!=j:
                    sum -= A[i][j] * x[j]     #Se va sustrayendo el valor actual de la matriz A en la posición i,j por el valor del vector X en la posición j cuando la coordenada de la matriz no corresponde a la diagonal principal y se le asigna a la variable sum

                x[i] = omega * sum + (1- omega) * old  # Se realiza Gauss Seidel calculan el promedio ponderado de los resultados obtenidos en las iteraciones anteriores y actuales, definidas en las variables sum y old, cuando el valor de omega es 1
                if centinela == 1 and x[i] != 0:
                   ea = abs((x[i] - old) / x[i]) * 100
                   if ea > tol:
                      centinela = 0

            print(iter)                     # Se imprime la iteración actual
            print(x)                        # Se imprime el valor de X actual
            iter += 1                       # Se incrementa la variable de iteración para realizar la siguiente
            if centinela == 1 or iter >= imax: # Evalua si la bandera centinela es 1 cuando el valor del error es menor o igual a la tolerancia o si el numero de iteraciones llega al valor máximo, haciendo que finalice el método de Gauss Seidel por medio de la directiva Break
                break
    return x                                # devuelve el valor de X 
```
