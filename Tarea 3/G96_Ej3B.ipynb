# Paso-2: Definición de la interpolación

```python
def InterpolacionNewton (x, y, x_inter):                          #Se Define la función de interpolación de Newton llamada InterpolacionNewton                
  n = len(x)                                                      #Se define n con el numero de posiciones de x 
 
  dfd = [[None for x in range(n)]  for x in range(n)]             #Se define la matriz vacia       
  yint = [None for x in range(n)]                                 #Se crea el vector vacio

  for i in  range(n):                                             #Se itera con un for la matriz de i hasta n definiendo un valor en cada iteración
    dfd[i][0] = y[i]                                   

  for j in range(1,n):                                            #Se itera con un for la primera columna
    for i in range(n-j):                                          #Se itera con un for de i hasta n-J
      dfd[i][j] = (dfd[i+1][j-1] - dfd[i][j-1]) / (x[i+j] - x[i]) #Se realiza la operación de la diferencia

  yint[0]  = dfd[0][0]                                              
  xterm = 1                                                        

  for k in range(1,n):                                            #Se itera con un for      
    xterm = xterm * (x_inter - x[k-1])                            #Se define xterm con la anterior operacion
    y_inter = yint[k-1] + dfd[0] [k] * xterm                      #Se define y_inter con la anterior operacion
    yint[k] = y_inter                                             #Se define yint en posicion k con y_inter                   

  return y_inter                                                  #Se retorna los valores en y_inter
```

# Paso-3: Llamar Función Para Interpolar 

```python
fxn = print(InterpolacionNewton(x, y, x_inter))   #Se imprime el resultado
```

#Paso-4: Grafica del Polinomio

```python
x_v = np.arange(min(x), max(x), 0.01)                     #Función para definir cada valor de x          
y_v = []                                                  #Se define una lista Vacia        

for i in range(len(x_v)):                                 #Recorremos los valores de x_v con un for  
  y_grafica = InterpolacionNewton(x, y, x_v[i])           #Se define y_grafica con la funcion de interpolación definida anteriormente      
  y_v.append(y_grafica)                                   #Se agrega y_grafica a y_v         
                                       
fxn = InterpolacionNewton (x, y, x_inter)                 #Se fxn haciendo un llamado nuevamente a la funcion InterpolacionNewton

plt.scatter(x, y, color="black")                          #A continuación de definen varios parametros para la grafica
plt.xlabel("X")                                   
plt.ylabel("Y")
plt.title("Interpolación de Newton")
plt.grid(True)
plt.plot(x_v, y_v,'r')
plt.scatter(x_inter, fxn, color='g')
```

# Paso-6: Función de interpolación de Lagrange

```python
def InterpolacionLagrange(x, y, x_inter):                       #Se Define la función de interpolación llamada InterpolacionLagrange    
  n = len(x)                                                    #Se define n con el numero de posiciones de x     
  suma = 0                                                      #Se define suma en 0     

  for i in  range(n):                                           #Se itera con un for de i hasta n para realizar unas operaciones en cada iteración segun su formula   
    producto = y[i]                                             #Se define producto con y en posicion de i segun cada iteracion    
    for j in  range(n):                                         #Se itera con un for cada fila    
      if i !=j:                                                 #Si i es diferente de j entra en el if    
       producto = producto * (x_inter - x[j]) / (x[i] - x[j])   #Realiza una operacion segun la formula y la almacena en producto    

    suma = suma + producto                                      #Se suma producto a la variable suma   
  y_inter = suma                                                #Se guarda la suma en y_inter
  
  return y_inter                                                #Se retorna los valores almacenados en y_inter
```

# Paso-7: Llamada a la función para interpolar

```python
fxn = print(InterpolacionLagrange(x, y, x_inter))  #Se imprime el resultado
```

# Paso-8: Generar Grafica

```python
x_v = np.arange(min(x), max(x), 0.01)                   #Función para definir cada valor de x            
y_v = []                                                #Se define una lista Vacia            

for i in range(len(x_v)):                               #Recorremos los valores de x_v con un for                
  y_grafica = InterpolacionLagrange(x, y, x_v[i])       #Se define y_grafica con la funcion de interpolación definida anteriormente             
  y_v.append(y_grafica)                                 #Se agrega y_grafica a y_v            
                                       
fxn = InterpolacionLagrange (x, y, x_inter)             #Se fxn haciendo un llamado nuevamente a la funcion InterpolacionLagrange

plt.scatter(x, y, color="black")                        #A continuación de definen varios parametros para la grafica
plt.xlabel("X")                                   
plt.ylabel("Y")
plt.title("Interpolación de Lagrange")
plt.grid(True)
plt.plot(x_v, y_v,'r')
plt.scatter(x_inter, fxn, color='g')
```
