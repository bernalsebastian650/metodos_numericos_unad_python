#**M√©todo de Jacobi**  
Es un m√©todo iterativo, usado para resolver sistemas de ecuaciones lineales del tipo. El algoritmo toma su nombre del matem√°tico alem√°n Carl Gustav Jakob Jacobi. El m√©todo de Jacobi consiste en usar f√≥rmulas como iteraci√≥n de punto fijo.


Un m√©todo iterativo con el cual se resuelve el sistema lineal A x = b comienza con una aproximaci√≥n inicial x (0) a la soluci√≥n x y genera una sucesi√≥n de vectores x (k) que converge a x. Los m√©todos iterativos traen consigo un proceso que convierte el sistema A x = b en otro equivalente de la forma x = T x + c para alguna matriz fija T y un vector c.


Luego de seleccionar el vector inicial x (0) la sucesi√≥n de los vectores de la soluci√≥n aproximada se genera calculando:

X (k) = Tx(k-1) + c 

Para cada k = 1, 2,3,...


Su formula general es: $$ x_{i}^{(k)} = -{\frac{\ 1}{a_{ii}}= \{\sum_{j=i}^{n}a_{ij}x_{j}^{(k-1)}}\} + \frac{\ b_{i}}{a_{ii}} $$  para cada i, 1<=i<=n

Comprobaci√≥n de diagonal dominante de la matriz A

```python
n = len(b)                                                                  # n toma el valor de la longitud del vector b
count = 0                                                                   # Contador
for i in range(n):                                                          # Se va a iterar en el rango de i hasta 10
    Aii = A[i][i]                                                           # Itera en la posicion i i de la matriz A
    suma = 0                                                                # La variable suma se define en cero
    for j in range(n):                                                      # Se va a iterar nuevamente en el rango de j hasta 10
        if i != j:                                                          # Determina si i es diferente de j 
           suma = suma + abs(A[i][i])                                       # Se realiza la suma absoluta de las posiciones i i segun su iteracion en cada fila
    if Aii > suma:                                                          # Determina Aii es mayor que la varible suma
       print("La fila", i, "es DD")                                         # imprime un mensaje indicando que i es diagonal dominante
       count += 1                                                           # Se le suma 1 al contador
    else:                                                                   # De lo contrario
       print("La fila", i, "No es DD")                                      # imprime un mensaje indicando que i no es diagonal dominante
       break                                                                # break finaliza la iteracion
if count == n:                                                              # Determina que el contador sea igual a n
    print("La matriz es DD y el m√©todo iterativo convergira.")              # imprime un mensaje he indicando que convergira
else:                                                                       # De lo contrario
    print("La matriz no es DD y el m√©todo iterativo puede que no converja") # Imprime un mensaje he indicando que no converja.
  
```

# $$ PASO-4 $$

```python
#Valores iniciales
tol = 0.000001                                       #Valor de la tolerancia de error
error = 100000                                       #Valor inicial de la norma                    De error = 3          error = 16
nitermax = 100                                       #N√∫mero m√°ximo de iteraciones
n = len(A)                                           #Determinador del tama√±o de A
x = np.zeros(n)                                      #Vector inicial de ceros  -----  x = [1,1,2,1,1,-3,1,1,1,1,1]
xl = np.zeros(n)                                     #Vector auxiliar de ceros
niter = 0                                            #Contador inicial de iteraciones
Niter = []
Error = []

while niter < nitermax and error > tol:              #Ciclo a iterar
    for i in range(n):                               # i es para recorrer las columnas
        d = b[i]
        for j in range(n):                           # j para recorrer las filas 
            if i != j:
              d = d - A[i][j] * x[j]
        x[i] = d / A[i][i]
    
    error = np.linalg.norm(x-xl)                     # Calculo de la norma vectorial
    xl = np.copy(x)                                  # Actualiza vector soluci√≥n iteraci√≥n anterior
    
    print("Iteraci√≥n:", niter)
    print(x)
    print("Error en la iteraci√≥n", niter, "es:", error)

    niter += 1
    Niter.append(niter)
    Error.append(error)
```

# Programaci√≥n M√©todo de Gauss - Seidel

# $$PASO-1 $$

```python
def SOR(A,b, x, imax, ea, omega):                     #Definicion de la funci√≥n SOR y los par√°metros requeridos
    n = len(x)                                        #n se le asigna la longitud del vector de ceros x
    for i in range(n):                                #Se itera con un for para asignar cada valor de la matriz A a la variable dummy
        dummy = A[i][i]         
        for j in range(n):                            #Se itera con un segundo for para recorrer cada valor de la matriz A dividirlo en dummy
            A[i][j] = A[i][j] / dummy

        b[i] = b[i] / dummy                           #Se itera en el vector B para divir su valor en dummy

    for i in range(n):                                #Se itera con un for para asignar los nuevos valores del vector b a la variable sum
        sum = b[i]
        for j in range(n):                            #Se itera con un for para ir sustrayendo el valor actual de la A en la posici√≥n i,j por el valor de x en la posici√≥n j
            if i != j:
              sum -= A[i][j] * x[j]
        x[i] = sum                                    #En cada iteracion se actualiza el valor de sum en la posici√≥n i del vector X

    iter = 1                                          #La variable iter se le asigna 1 

    while iter <= imax or ea > tol:                   #Si la iter es menor o igual al numero deseado de iteraciones o mientras el error sea mayor que la tolerancia va a iterar
        centinela = 1
        for i in range(n):                            #Se asignan los valores actuales de x en la variable old y del vector b en la variable sum
            old = x[i]
            sum = b[i]
            for j in range(n):                        #Se itera con un for para asignar los nuevos valores del vector b a la variable sum
                if i!=j:
                    sum -= A[i][j] * x[j]             #Se sustrae el valor actual de la matriz A en la posici√≥n i,j multiplicado por valor del vector x en la posici√≥n j

                x[i] = omega * sum + (1- omega) * old #Se aplica Gauss Seidel calculando el promedio ponderado de los resultados obtenidos en las iteraciones anteriores y actuales, definidas en las variables sum y old
                if centinela == 1 and x[i] != 0:
                   ea = abs((x[i] - old) / x[i]) * 100
                   if ea > tol:
                      centinela = 0

            print(iter)                               #Se muestra en pantalla la interaci√≥n actual
            print(x)                                  #Se muestra en pantalla el valor de X actual
            iter += 1                                 #Se incrementa 1 la iteracion
            if centinela == 1 or iter >= imax:        #Si centinela es igual a 1 o el valor la iteracion es mayor o igual a la iteracion hace que termine la iteracion con Break
                break
    return x                                          
```
